```python
points_1819 = pd.read_csv(r"C:\Users\Owner\OneDrive\Documents\data_science\points_1819.csv", encoding='Windows-1252')
```

```python
# Creating a 'Player' column to match up with the 'Player' column in the previous dataframe
points_1819['Player'] = np.where(points_1819['second_name']=='', points_1819['first_name'], points_1819['first_name'] + " " + points_1819['second_name'])
```

```python
# Dropping the two name columns after creating the 'Player' column
points_1819.drop(['first_name', 'second_name'], axis=1, inplace=True)
```

```python
# Making 'Player' column the index
points_1819.index = points_1819['Player']
points_1819.drop(['Player'], axis=1, inplace=True)
```

```python
# Removing whitespace at the beginning and end of the values
points_1819.index.str.strip()
```

```python
# Renaming index values with unknown characters (only 12 so not difficult to do manually)
features.rename(index={'Marko Arnautovi?':'Marko Arnautovic', 'Asmir Begovi?':'Asmir Begovic', 'Petr ?ech':'Petr Cech',
                          '?ukasz Fabia?ski':'Lukasz Fabianski', '?lkay Gündo?an':'Ilkay Gündogan', 'Mateo Kova?i?':'Mateo Kovacic',
                          'Nemanja Mati?':'Nemanja Matic', 'Luka Milivojevi?':'Luka Milivojevic', 'Yoshinori Mut?':'Yoshinori Muto',
                          'Ça?lar Söyüncü':'Çaglar Söyüncü', 'Mat?j Vydra':'Matej Vydra'}, inplace=True)
```

```python
# Joining the points data onto the season data 
final_feats = features.join(points_1819)
final_feats
```

```python
# Cutting 'Pos' value lengths to two characters
final_feats['Pos'] = np.where(final_feats['Pos'].str.len() == 4, final_feats['Pos'].str[2:], final_feats['Pos'])
```

```python
# Creating a separate dataframe for goalkeepers
gk_df = final_feats[final_feats.Pos == 'GK']
gk_df
```

```python
final_feats.drop(final_feats[final_feats.Pos == "GK"].index, axis=0, inplace=True)
```

```python
final_feats.columns
```

```python
# Feature engineering
# Adding total number of clearances, blocks and interceptions
final_feats['CBI'] = final_feats['Blocks'] + final_feats['Int'] + final_feats['Clr']
# Penalty kicks missed
final_feats['PK_miss'] = final_feats['PKatt'] - final_feats['PK']
# Number of net successful tackles
final_feats['succ_tackle'] = final_feats['TklW'] - (final_feats['Tkl'] - final_feats['TklW'])
```
